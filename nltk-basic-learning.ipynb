```python
import nltk
```

```python
text1="Trying to learn NLTK for Text preprocessing. It will help me devoloping some good system."
print(text1)
```

```python
text1.split(' '), len(text1.split(' '))

#We are getting full stop with the words here
```

```python
nltk.word_tokenize(text1),  len(nltk.word_tokenize(text1))
#here we re getting the full stop  like a separate token
```

```python
#now lets tokenize sentences 
nltk.sent_tokenize(text1) , len(nltk.sent_tokenize(text1))
```

# Normalizing And Steaming

```python
#Porter stemmer. The Porter stemming algorithm is a process for removing suffixes from words in English
porter = nltk.PorterStemmer()
text_lower = text1.lower()
words=nltk.word_tokenize(text_lower)
print(words)
```

```python
[porter.stem(w) for w in words]
```

# Lemmatization

```python
udhr=nltk.corpus.udhr.words('English-Latin1')
#it is a function in the NLTK library that retrieves a list of words from the Universal Declaration of Human Rights (UDHR) corpus in various languages.
print(udhr)
```

```python
udhr[:30]
```

```python
[porter.stem(w) for w in udhr]
#we trim the words but these seems meaning less now, ok lets get to lemmatization next
```
